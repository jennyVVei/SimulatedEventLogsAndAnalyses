```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
db_path = "Dirigo.sqlite"
conn = sqlite3.connect(db_path)
```

```python
df_cargo = pd.read_sql_query("SELECT * FROM OCEL_Cargo",conn)
df_truck = pd.read_sql_query("SELECT * FROM OCEL_Truck",conn)
df_pickup = pd.read_sql_query("SELECT * FROM OCEL_PickupPLan",conn)
df_e2o = pd.read_sql_query("SELECT * FROM OCEL_E2O",conn)
df_o2o = pd.read_sql_query("SELECT * FROM OCEL_O2O", conn)
df_events = pd.read_sql_query("SELECT * FROM OCEL_Events",conn)

conn.close()
```

```python
combined_events_e2o = pd.merge(df_events, df_e2o, on="Event_id", how= 'inner')
combined_events_e2o
```

```python
df_truck
```

```python
truck_merged_df = pd.merge(combined_events_e2o, df_truck, on=['Object_id', 'Timestamp'],how = 'outer')
truck_merged_df
```

```python
pp_merged_df = pd.merge(truck_merged_df,df_pickup, on=['Object_id', 'Timestamp'],how = 'outer')
pp_merged_df
```

```python
lodge_merged_filtered = pp_merged_df[pp_merged_df['Activity']=='Lodge Pickup Plan']
lodge_merged_filtered
```

```python
lodge_merged_filtered['Cargo ID'] = lodge_merged_filtered['Cargo ID'].fillna(lodge_merged_filtered['CargoID'])
pcps = lodge_merged_filtered['Pickup Plan ID'].isna() & lodge_merged_filtered['Object_id'].str.startswith('Pcp')
lodge_merged_filtered.loc[pcps, 'Pickup Plan ID'] = lodge_merged_filtered.loc[pcps, 'Object_id']
lodge_merged_filtered = lodge_merged_filtered[~lodge_merged_filtered['Object_id'].str.startswith('Cr', na=False)]
lodge_merged_filtered.drop(columns=['CargoID','Ocel_changed_field_y'], inplace=True)
lodge_merged_filtered.rename(columns={'Ocel_changed_field_x':'Ocel_changed_field'}, inplace=True)
lodge_merged_filtered
```

```python
non_lodge_merged_filtered = pp_merged_df[pp_merged_df['Activity']!='Lodge Pickup Plan']
non_lodge_merged_filtered.drop(columns=['CargoID','Ocel_changed_field_y'], inplace=True)
non_lodge_merged_filtered.rename(columns={'Ocel_changed_field_x':'Ocel_changed_field'}, inplace=True)
```

```python
lodge_merged_trucks = pd.concat([lodge_merged_filtered,non_lodge_merged_filtered]).reset_index(drop=True)
lodge_merged_trucks
```

```python
cargo_merged_rest = pd.merge(lodge_merged_trucks, df_cargo, on=['Object_id', 'Timestamp'], how='outer')
cargo_merged_rest
```

```python
cargo_merged_rest['Ocel_changed_field_x'] = cargo_merged_rest['Ocel_changed_field_x'].fillna(cargo_merged_rest['Ocel_changed_field_y'])
cargo_merged_rest.drop(columns=['Ocel_changed_field_y'], inplace=True)
cargo_merged_rest.rename(columns={'Ocel_changed_field_x':'Ocel_changed_field'}, inplace=True)
cargo_merged_rest
```

```python
lodge_ts = cargo_merged_rest[cargo_merged_rest['Activity'] == 'Lodge Pickup Plan']['Timestamp'].unique()
len(lodge_ts)
```

```python
df_update_part = cargo_merged_rest[(cargo_merged_rest['Activity'] == 'Lodge Pickup Plan') | ((cargo_merged_rest['Event_id'].isna()) & (cargo_merged_rest['Timestamp'].isin(lodge_ts)) & cargo_merged_rest['Object_id'].str.startswith('Cr'))].copy()
df_update_part
```

```python
df_rest = cargo_merged_rest.drop(df_update_part.index)
df_rest
```

```python
lodge_mask = df_update_part['Activity'] == 'Lodge Pickup Plan'
lodge_idxs = df_update_part[lodge_mask].index
filler_df = df_update_part[~lodge_mask] 
```

```python
for idx in lodge_idxs:
    ts = df_update_part.at[idx, 'Timestamp']
    matching_rows = filler_df[filler_df['Timestamp'] == ts]
    
    for _, mrow in matching_rows.iterrows():
        for col in cargo_merged_rest.columns:
            val = df_update_part.at[idx, col]
            if pd.isna(val) or val in ['', 'None', 'nan', 'NaN']:
                if not pd.isna(mrow[col]) and mrow[col] not in ['', 'None', 'nan', 'NaN']:
                    df_update_part.at[idx, col] = mrow[col]
```

```python
df_update_part[lodge_mask]
```

```python
df_final = pd.concat([df_update_part[lodge_mask], df_rest], ignore_index=True)
df_final
```

```python
df_final.loc[(df_final['Ocel_changed_field'].isna()) & df_final['LPT'].notna(), 'Activity'] = 'Review the Truck Information'
df_final[df_final['Activity']== 'Review the Truck Information']
```

```python
df_final.loc[(df_final['Ocel_changed_field'].isna()) & df_final['Cargo Type'].notna(), 'Activity'] = 'Review the Cargo Information'
df_final[df_final['Activity']== 'Review the Cargo Information']
```

```python
assign_trucks_df =df_final[df_final['Activity'] == 'Assign Truck']
assign_trucks_df
```

```python
assign_trucks_df =df_final[df_final['Activity'] == 'Assign Truck']
assign_trucks_df=assign_trucks_df[assign_trucks_df['Object_id'].str.startswith('tr')]
assign_trucks_df
```

```python
other_activities_df = df_final[df_final['Activity'] != 'Assign Truck']
other_activities_df
```

```python
def merge_group_rows_with_field(group):
    merged_row = group.iloc[0].copy()
    for col in group.columns:
        if col not in ['Event_id', 'Activity', 'Timestamp', 'Object_id', 'Ocel_changed_field']:
            non_null_values = group[col].dropna().unique()
            if len(non_null_values) > 0:
                merged_row[col] = non_null_values[0]

    # Populate the value from 'ocel_changed_field' into the corresponding column
    for _, row in group.iterrows():
        field = row['Ocel_changed_field']
        if field in group.columns and pd.notna(row[field]):
            merged_row[field] = row[field]

    return merged_row
```

```python
assigned = assign_trucks_df.groupby(['Event_id', 'Timestamp']).apply(merge_group_rows_with_field).reset_index(drop=True)
assigned
```

```python
all_xes = pd.concat([assigned, other_activities_df]).sort_values(by='Timestamp').reset_index(drop=True)
all_xes
```

```python
all_xes.loc[(all_xes['Ocel_changed_field'] == 'Truck Status') & all_xes['Activity'].isna(), 'Activity'] = 'Prepare Reassignment'
all_xes[all_xes['Activity']== 'Prepare Reassignment']
```

```python
crs = all_xes['Object_id'].str.startswith('Cr')
all_xes.loc[crs, 'Cargo ID'] = all_xes.loc[crs, 'Object_id']
all_xes
```

```python
Pcps = all_xes['Object_id'].str.startswith('Pcp')
all_xes.loc[Pcps, 'Pickup Plan ID'] = all_xes.loc[Pcps, 'Object_id']
all_xes
```

```python
all_xes['Truck ID'] = all_xes['Object_id'].where(all_xes['Object_id'].str.startswith('tr'),None)
all_xes
```

```python
trucks = all_xes[all_xes['Object_id'].str.startswith('tr')]
non_trucks = all_xes[~all_xes['Object_id'].str.startswith('tr')]
len(non_trucks)
```

```python
truck_ordered = trucks.sort_values(by=['Object_id', 'Timestamp']).reset_index(drop=True)

# Reinitialize the 'pickup' column
truck_ordered['pickup'] = None

# Initialize variables to track pickups
pickup_id = 0
assigning_pickup = False
current_truck = None

# Iterate over the rows to assign pickup IDs
for idx, row in truck_ordered.iterrows():
    activity = row['Activity']
    truck_id = row['Object_id']

    if activity == "Assign Truck":
        pickup_id += 1
        assigning_pickup = True
        current_truck = truck_id
        truck_ordered.at[idx, 'pickup'] = pickup_id

    elif assigning_pickup and truck_id == current_truck:
        truck_ordered.at[idx, 'pickup'] = pickup_id
        if activity in ["Evaluate the Truck Exit"]:
            assigning_pickup = False
```

```python
truck_ordered
```

```python
pp_cr_ids = truck_ordered.dropna(subset=['pickup']).drop_duplicates(subset=['pickup'])[['pickup', 'Pickup Plan ID','Cargo ID']]
pp_cr_ids
```

```python
alltrucks_df = pd.merge(truck_ordered, pp_cr_ids, on='pickup', how='outer', suffixes=('', '_filled'))
alltrucks_df
```

```python
alltrucks_df['Cargo ID'] = alltrucks_df['Cargo ID'].fillna(alltrucks_df['Cargo ID_filled'])
alltrucks_df.drop(columns=['Cargo ID_filled'], inplace=True)
alltrucks_df
```

```python
non_trucks['Pickup Plan ID_filled'] = None
groupPP = non_trucks['Pickup Plan ID_filled'].isna() & non_trucks['Pickup Plan ID'].notna()
non_trucks.loc[groupPP, 'Pickup Plan ID_filled'] = non_trucks.loc[groupPP, 'Pickup Plan ID']
non_trucks
```

```python
all_df = pd.concat([alltrucks_df, non_trucks], ignore_index=True)
all_df
```

```python
all_df.drop(columns=['Ocel_changed_field','Pickup Plan ID'],inplace=True)
all_df.rename(columns={'Pickup Plan ID_filled': 'Pickup Plan ID'}, inplace=True)
```

```python
all_df['Cargo ID'] = all_df['Cargo ID'].replace(['', ' '], pd.NA)
all_df['Cargo ID'] = all_df.groupby('pickup')['Cargo ID'].transform(lambda x: x.ffill().bfill())
all_df
```

```python
crs1 = all_df['Object_id'].str.startswith('Cr')
all_df.loc[crs1, 'Cargo ID'] = all_df.loc[crs1, 'Object_id']
all_df
```

```python
all_df.to_csv('all.csv',index=False)
```
